```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('data/StudentsPerformance.csv')
df
```

```python
df.isnull().sum()/len(df)
```

```python
df=df.rename(columns={'race/ethnicity':'group','parental level of education':'parents_edu','test preparation course':'course','math score':'maths','reading score':'reading','writing score':'writing'})
```

```python
df['gender']=df['gender'].map({'female':1,'male':0})
df['group']=df['group'].map({'group A':1,'group B':2,'group C':3,'group D':4,'group E':5})
df['course']=df['course'].map({'completed':1,'none':0})
```

```python
df['percentage']=(df['maths']+df['reading']+df['writing'])/3
df
```

```python
female_above_85=((df['gender']==1) & (df['percentage']>85)).sum()/(df['gender']==1).sum()*100
male_above_85=((df['gender']==0) & (df['percentage']>85)).sum()/(df['gender']==0).sum()*100
print(f'''percentage of female student in first class:{female_above_85}
percentage of male student in first class:{male_above_85}''')
```

```python
female_passed=((df['gender']==1) & (df['percentage']>40)).sum()/(df['gender']==1).sum()*100
male_passed=((df['gender']==0) & (df['percentage']>40)).sum()/(df['gender']==0).sum()*100
print(f'''percentage of female students passed:{female_passed}
percentage of male student passed:{male_passed}''')
```

```python
course_student_percentage=((df['course']==1) & (df['percentage']>40)).sum()/(df['course']==1).sum()*100
noncourse_student_percentage=((df['course']==0) & (df['percentage']>40)).sum()/(df['course']==0).sum()*100
print(f'''possibility of passing for students completing test preperation course:{course_student_percentage}
possibility of passing for students not doing test preperation course:{noncourse_student_percentage}
hence,''')
if course_student_percentage>noncourse_student_percentage:
    print("possibility of passing for students completing test preperation course is more than others")
else:
    print("possibility of passing for students not doing test preperation course is more than ones who completed")
```

```python
course_student_topper_percentage=((df['course']==1) & (df['percentage']>85)).sum()/(df['course']==1).sum()*100
noncourse_student_topper_percentage=((df['course']==0) & (df['percentage']>85)).sum()/(df['course']==0).sum()*100
print(f'''possibility of being topper for students completing test preperation course:{course_student_topper_percentage}
possibility of being topper for students not doing test preperation course:{noncourse_student_topper_percentage}
hence,''')
if course_student_topper_percentage>noncourse_student_topper_percentage:
    print("possibility of being topper for students completing test preperation course is more than others")
else:
    print("possibility of being topper for students not doing test preperation course is more than ones who completed")
```

```python
edu_order = {
    'some high school': 1,
    'high school': 2,
    'some college': 3,
    "associate's degree": 4,
    "bachelor's degree": 5,
    "master's degree": 6
}
df['parents_edu_level'] = df['parents_edu'].map(edu_order)
```

```python
def classify_edu(edu):
    if edu in ['some high school', 'high school']:
        return 'low'
    elif edu in ['some college', "associate's degree"]:
        return 'medium'
    else:
        return 'high'

df['edu_category'] = df['parents_edu'].apply(classify_edu)
```

```python
df.groupby('edu_category')['percentage'].mean()
```
